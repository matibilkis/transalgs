```python
import dask.dataframe as dd
from dask.diagnostics import ProgressBar
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = dd.read_csv("datos_SUBE/Datos_Vega.csv",encoding = "ISO-8859-1", sep=";")
```

```python
with ProgressBar():
    display(df.head())
```

```python
columns_keep = ["LONGITUD","LATITUD","FECHA_HORA","RAMAL","MODO"]
df_fil = df[columns_keep]


cond = "MODO" == "COL"
condition_modo = df_fil["MODO"] == "COL"
df_fil_modo = df_fil[condition_modo]

df_fil1 = df_fil_modo.drop(["MODO"], axis=1)
```

```python
par1 = df_fil1.get_partition(0)
with ProgressBar():
    dp1 = par1.compute()
```

```python
dp1["HORA"] = dp1["FECHA_HORA"].apply(lambda x: x[-2:])
dp1["FECHA"] = dp1["FECHA_HORA"].apply(lambda x: x[:-3])
dp1_cleaner = dp1.drop(["FECHA_HORA"],axis=1)
```

```python
dp1_cleaner.head(10)
```

```python
dp1_date_index = dp1_cleaner.set_index(["FECHA"])
```

```python
dp1_date_index
```

```python
from datetime import datetime

datetime_object = datetime.strptime('Jun 1 2005  1:33PM', '%b %d %Y %I:%M%p')
```

```python
convert_to_datetime = lambda x: datetime.strptime(x,"%d/%m/%Y")
```

```python
dp1_date_index["FECHA"].set_index(convert_to_datetime(dp1.index))
```
